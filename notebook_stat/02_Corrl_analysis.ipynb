```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_crypto_15 = pd.read_csv('../data/ALL_15_interpol.csv')
#all_crypto_15.dropna(inplace = True)


all_crypto_15.set_index(['open_time'],inplace =True)
all_crypto_15.index = pd.to_datetime(all_crypto_15.index)
all_crypto_15.head()
```

```python
all_crypto_15.loc['2020']
```

```python
all_crypto_15_norm = all_crypto_15 / all_crypto_15.iloc[0,:]
```

```python
all_crypto_15_norm.plot()
plt.legend(bbox_to_anchor = (1,1))
```

```python
pd.plotting.scatter_matrix(all_crypto_15,alpha = 0.1,figsize = (16,20))
```

```python
all_crypto_15_return = np.log(all_crypto_15).diff().dropna()
all_crypto_15_return.index = pd.to_datetime(all_crypto_15_return.index)
corr = all_crypto_15_return.corr()
```

```python
pd.plotting.scatter_matrix(all_crypto_15_return,alpha = 0.1,figsize = (16,20))
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(corr, cmap="Blues", annot=True)
```

```python
all_years = all_crypto_15_return.index.year.unique()
all_years
```

```python
all_crypto_15_return.columns
```

```python
high_correlation = ['close_AAVEUSDT', 'close_BCHUSDT', 'close_BNBUSDT',
       'close_BTCUSDT', 'close_ETHUSDT','close_ILVUSDT',
       'close_LTCUSDT', 'close_MKRUSDT', 'close_XMRUSDT','close_YFIUSDT']
```

```python
YEAR = '2022'
all_crypto_15_return = np.log(all_crypto_15[high_correlation]).loc[YEAR].diff().iloc[1:].dropna(axis = 1 )
all_crypto_15_return.index = pd.to_datetime(all_crypto_15_return.index)
corr = all_crypto_15_return.corr()

plt.figure(figsize = (10,10))
sns.heatmap(corr, cmap="Blues", annot=True)
plt.title(YEAR)
```

```python
YEAR = '2021'
all_crypto_15_return = np.log(all_crypto_15[high_correlation]).loc[YEAR].diff().iloc[1:].dropna(axis = 1 )
all_crypto_15_return.index = pd.to_datetime(all_crypto_15_return.index)
corr = all_crypto_15_return.corr()

plt.figure(figsize = (10,10))
sns.heatmap(corr, cmap="Blues", annot=True)
plt.title(YEAR)
```
